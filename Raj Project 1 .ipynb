##### 12. Write a function using only list comprehensions, no loops, to compute standard deviation. Using this function, calculate the standard deviation of each numeric column in both data sets. Add these to a list called `sd`.

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

```python
#Need help on this one, to fully understand 

def standarddev(df_column):
    num_var = df_column.size 
    mean = float(df_column.sum()/num_var) 
    return (sum((i-mean)**2 for i in df_column)/(num_var-1))**0.5
```

```python
sd = both_tests.drop("State",axis=1).apply(standarddev)
sd
```

##### 26. Execute a hypothesis test comparing the SAT and ACT participation rates. Use $\alpha = 0.05$. Be sure to interpret your results.

```python
## Using code, this one is still tough 

                        #The point estimate is the difference between the two
act_mean = both_tests['ACT Participation'].mean()
test_diff = act_mean - sat_mean
print("The point estimate is",test_diff)

#T test and p value 

t_stat, p_value = stats.ttest_ind(both_tests['ACT Participation'], both_tests['SAT Participation'])
print("The t statistic is {} and the p_value is {}".format(t_stat, p_value))
```

```python
#Since the point estimate is 25%, it means there is a difference between SAT and ACT participation of that amount 
#The p value is <.05 so we reject the null hypothesis. And there is a mean difference between SAT/ACT participation. 
```

##### 27. Generate and interpret 95% confidence intervals for SAT and ACT participation rates.

```python
def  confidence(sample, sd=0.95): #sample = both_tests 
    
    
    zscore = stats.norm.ppf(1-(1-sd)/2) 
    
    low_ci = sample.mean() - zscore*sample.sem()

    high_ci = sample.mean() + zscore*sample.sem()
    

    return (low_ci, high_ci)
```

```python
confidence(both_tests['SAT Participation']) #95% fall between .30 and .49 
```

```python
confidence(both_tests['ACT Participation']) #95% fall between .56 and .74
```
